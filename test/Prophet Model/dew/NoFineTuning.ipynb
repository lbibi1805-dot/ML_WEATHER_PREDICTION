```python
# Required imports
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from prophet import Prophet
import matplotlib.pyplot as plt

# Load the dataset
file_path = '../../../dataset/data.csv'
data = pd.read_csv(file_path)

# Prepare the data for Prophet (for dew
data['ds'] = pd.to_datetime(data['datetime'], format='%m/%d/%Y')
dew_data = data[['ds', 'dew']]  # Rename dewfor Prophet
dew_data.rename(columns={'dew': 'y'}, inplace=True)

# Train-test split (80% training, 20% testing)
train_size = int(0.8 * len(dew_data))
train_data = dew_data[:train_size]  # 80% for training
test_data = dew_data[train_size:]   # 20% for validation

# Initialize and fit the Prophet model for dew_model = Prophet(yearly_seasonality=True)
dew_model = Prophet(yearly_seasonality=True)
dew_model.fit(train_data)

# Print the hyperparameters of the model
print("Model Hyperparameters:")
print(f"Yearly Seasonality: {dew_model.yearly_seasonality}")
print(f"Weekly Seasonality: {dew_model.weekly_seasonality}")
print(f"Daily Seasonality: {dew_model.daily_seasonality}")
print(f"Changepoint Prior Scale: {dew_model.changepoint_prior_scale}")
print(f"Seasonality Prior Scale: {dew_model.seasonality_prior_scale}")
print(f"Seasonality Mode: {dew_model.seasonality_mode}")

# Predict on the test set (validation set)
test_forecast = dew_model.predict(test_data[['ds']])

# Calculate RÂ², RMSE, and MAE
y_true = test_data['y'].values  # Actual dew
y_pred = test_forecast['yhat'].values  # Predicted dew

# Calculate the RÂ² score
r2 = r2_score(y_true, y_pred)

# Calculate the RMSE
rmse = mean_squared_error(y_true, y_pred, squared=False)

# Calculate the MAE
mae = mean_absolute_error(y_true, y_pred)

# Display the performance metrics
print(f"RÂ² Score: {r2:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"MAE: {mae:.3f}")

# Generate future dates (from 21/08/2024 to 31/12/2024)
future_dates = dew_model.make_future_dataframe(periods=len(pd.date_range('2024-08-21', '2024-12-31')), freq='D')

# Predict future dew
dew_forecast = dew_model.predict(future_dates)

# Display and save the predicted dew
dew_forecast[['ds', 'yhat']].tail()
dew_forecast[['ds', 'yhat']].to_csv('dew_forecast_2024.csv', index=False)

# Plot the forecasted dew
dew_model.plot(dew_forecast)
plt.title("dewForecast (2024)")
plt.show()
```
