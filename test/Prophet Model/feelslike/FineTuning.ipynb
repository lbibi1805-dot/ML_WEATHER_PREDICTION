```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = '../../../dataset/data.csv'  # Adjust the path to your dataset
data = pd.read_csv(file_path)

# Convert datetime to the correct format and rename for Prophet
data['ds'] = pd.to_datetime(data['datetime'], format='%m/%d/%Y')

# Prepare the target variable (feelslike) and features for Prophet
data['y'] = data['feelslike']

# Select the relevant features
features = ['temp', 'tempmax', 'tempmin', 'feelslikemax', 'feelslikemin', 'dew', 'sealevelpressure', 'cloudcover']

# Hyperparameters to tune
param_grid = {
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

best_score = float('-inf')  # To track the best model
best_params = None  # To store the best hyperparameters

# Step 1: Grid search over the hyperparameters
for changepoint_prior in param_grid['changepoint_prior_scale']:
    for seasonality_prior in param_grid['seasonality_prior_scale']:
        for seasonality_mode in param_grid['seasonality_mode']:
            # Initialize Prophet model with current hyperparameters
            feelslike_model = Prophet(
                changepoint_prior_scale=changepoint_prior,
                seasonality_prior_scale=seasonality_prior,
                seasonality_mode=seasonality_mode
            )

            # Add regressors
            for feature in features:
                feelslike_model.add_regressor(feature)

            # Fit the model
            feelslike_model.fit(data[['ds', 'y'] + features])

            # Generate predictions on the historical data
            historical_forecast = feelslike_model.predict(data[['ds'] + features])
            y_true = data['y'].values  # Actual feelslike temperatures
            y_pred = historical_forecast['yhat'].values  # Predicted feelslike temperatures

            # Calculate R², RMSE, and MAE
            r2 = r2_score(y_true, y_pred)
            rmse = mean_squared_error(y_true, y_pred, squared=False)
            mae = mean_absolute_error(y_true, y_pred)

            # Track the best model based on R² score
            if r2 > best_score:
                best_score = r2
                best_params = {
                    'changepoint_prior_scale': changepoint_prior,
                    'seasonality_prior_scale': seasonality_prior,
                    'seasonality_mode': seasonality_mode
                }

            # Print performance for the current model
            print(f"Changepoint: {changepoint_prior}, Seasonality: {seasonality_prior}, Mode: {seasonality_mode}")
            print(f"R²: {r2:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}\n")

# Step 2: Report the best hyperparameters
print(f"Best Hyperparameters: {best_params}")
print(f"Best R² Score: {best_score:.3f}")

# Step 3: Train the best model on the full dataset using the best hyperparameters
best_model = Prophet(
    changepoint_prior_scale=best_params['changepoint_prior_scale'],
    seasonality_prior_scale=best_params['seasonality_prior_scale'],
    seasonality_mode=best_params['seasonality_mode']
)

# Add the same regressors
for feature in features:
    best_model.add_regressor(feature)

# Fit the best model
best_model.fit(data[['ds', 'y'] + features])

# Step 4: Predict future feelslike temperatures
future_dates = best_model.make_future_dataframe(periods=len(pd.date_range('2024-08-21', '2024-12-31')), freq='D')
historical_averages = data.groupby(data['ds'].dt.dayofyear)[features].mean()
future_features = pd.DataFrame({
    col: [historical_averages.loc[dayofyear if dayofyear in historical_averages.index else dayofyear - 1, col]
          for dayofyear in future_dates['ds'].dt.dayofyear]
    for col in features
})
future_df = pd.concat([future_dates, future_features], axis=1)
feelslike_forecast = best_model.predict(future_df)

# Step 5: Calculate confidence intervals (using standard deviation)
feelslike_predictions = feelslike_forecast['yhat'].values
std_dev = np.std(feelslike_predictions)
lower_bound = feelslike_predictions - 1.96 * std_dev
upper_bound = feelslike_predictions + 1.96 * std_dev

# Step 6: Create a DataFrame to store the predictions and confidence intervals
future_predictions_df = pd.DataFrame({
    'Date': future_dates['ds'],
    'Predicted Feelslike': feelslike_predictions,
    'Lower Bound': lower_bound,
    'Upper Bound': upper_bound
})

# Save the predicted values to a CSV file
future_predictions_df.to_csv('finetuning.csv', index=False)
print("Predictions saved to 'finetuning.csv'")

# Step 7: Plot the predicted feelslike temperatures with confidence intervals
plt.figure(figsize=(10, 6))
plt.plot(future_predictions_df['Date'], future_predictions_df['Predicted Feelslike'], label='Predicted Feelslike', color='blue', linewidth=2)
plt.fill_between(future_predictions_df['Date'], future_predictions_df['Lower Bound'], future_predictions_df['Upper Bound'], color='blue', alpha=0.2, label='Confidence Interval (95%)')
plt.scatter(data['ds'], data['feelslike'], color='black', label='Actual Feelslike', alpha=0.5)  # Plot actual values as black dots
plt.xlabel('Date')
plt.ylabel('Feelslike Temperature')
plt.title('Feelslike Temperature Predictions with Confidence Intervals for 2024')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
